### ID Mappings

```python
admission_type_id_df = ids_mapping_df[0:8]
discharge_decomposition_id_df = ids_mapping_df[10:40]
admission_source_id_df = ids_mapping_df[42:66]
```

```python
def map_id_to_description(id_number, id_type):
    """
    This function maps the id and the id_type to
    a description.
    :param id_number: Id number of type String
    :param id_type: Type of the ID: ["admission_type", 
                                    "admission_source", 
                                    "discharge_decomposition"]
    :returns descrition: String containing the Description of the
    admission type.
    """
    if id_type == "admission_type":
        description = admission_type_id_df[admission_type_id_df["admission_type_id"]==id_number]["description"].values[0]
    elif id_type == "admission_source":
        description = admission_source_id_df[admission_source_id_df["admission_type_id"]==id_number]["description"].values[0]
    elif id_type == "discharge_decomposition":
        description = discharge_decomposition_id_df[discharge_decomposition_id_df["admission_type_id"]==id_number]["description"].values[0]
    else:
        raise ValueError("No such entry found")
        
    return description
```

### Diagnonsis Binning
Here the diag_1, diag_2 and diag_3 are binned into groups int a new column called as diagnostic_tests. The main reason for this is to reduce the number of features. The binning happens according to the diagnostic type provided in this image
![title](original_data/diag_bins.png)

```python
def determine_diag_map(icd9_code):
        """
        This function determines the diagnotive map
        depending on the table above. It Assigns a 
        category referred by the group name in lower case
        :param icd9_codes: String
        """
        if icd9_code.find("E") != -1 or \
            icd9_code.find("V") != -1:
            return "other"
        elif icd9_code.find("?") != -1:
                return None
        else:
            icd9_code = float(icd9_code)
            if 390<=icd9_code<=459 or \
                icd9_code == 785:
                return "circulatory"
            elif 460<=icd9_code<=519 or \
                icd9_code == 786:
                return "respiratory"
            elif 520<=icd9_code<=579 or \
                icd9_code == 787:
                return "digestive"
            elif 250<=icd9_code<251 or \
                icd9_code == 787:
                return "diabetes"
            elif 800<=icd9_code<=999:
                return "injury"
            elif 710<=icd9_code<=739:
                return "musculoskeletal"
            elif 580<=icd9_code<=629 or \
                icd9_code == 788:
                return "genitourinary"
            elif 140<=icd9_code<=239:
                return "neoplasms"
            else:
                return "other"
```

```python
possible_diagnosis = [
                        "circulatory",
                        "respiratory",
                        "digestive",
                        "diabetes",
                        "injury",
                        "musculoskeletal",
                        "genitourinary",
                        "neoplasms",
                        "other"
                     ]
```

```python
for diagnosis_type in possible_diagnosis:
    diabatic_data_df[diagnosis_type] = ""
```

```python
diabatic_data_df["diag_1_binned"] = list(map(determine_diag_map, diabatic_data_df["diag_1"].values))
diabatic_data_df["diag_2_binned"] = list(map(determine_diag_map, diabatic_data_df["diag_2"].values))
diabatic_data_df["diag_3_binned"] = list(map(determine_diag_map, diabatic_data_df["diag_3"].values))
```

```python
diagnosis_df = diabatic_data_df[["diag_1_binned", 
                                 "diag_2_binned", 
                                 "diag_3_binned"]].stack().str.get_dummies().sum(level=0)
diagnosis_df[diagnosis_df>1]=1
```
